17.แสดงชื่อประเทศที่เข้าร่วมโอลิมปิกฤดูร้อนมากที่สุด (ระวัง!! อาจมีหลายประเทศ)

18.จำนวนประเทศที่เข้าร่วมโอลิมปิกฤดูหนาวบ่อยที่สุดมีกี่ประเทศ

19.จำนวนเหรียญทั้งหมดที่แจกไปในโอลิมปิกฤดูหนาวรวมแล้วมีกี่เหรียญ

```python
medals[['WinterGold','WinterSilver','WinterBronze']].sum().sum()
```

20.หาค่าเฉลี่ยของจำนวนเหรียญทองฤดูหนาวและฤดูร้อนต่อประเทศ

```python
print("Average of winter gold:",medals['WinterGold'].mean())
print("Average of summer gold:",medals['SummerGold'].mean())
```

21.แสดงจำนวนเหรียญทั้งหมด(รวมฤดูร้อนและฤดูหนาว)ที่แต่ละประเทศได้

```python
medals['TotalMedals'] = medals[['WinterGold', 'WinterSilver', 'WinterBronze',
                                 'SummerGold', 'SummerSilver', 'SummerBronze']].sum(axis=1)

medals[['Country', 'TotalMedals']]
```

22.แสดงประเทศที่ได้เหรียญรวมทุกประเภทของโอลิมปิกฤดูหนาวมากกว่าเหรียญรวมโอลิมปิกฤดูร้อนให้คอลัมน์ของตารางนี้แสดงชื่อประเทศเหรียญรวมฤดูร้อนเหรียญรวมฤดูหนาว

```python
medals['TotalWinterMedals'] = medals[['WinterGold', 'WinterSilver', 'WinterBronze']].sum(axis=1)
medals['TotalSummerMedals'] = medals[['SummerGold', 'SummerSilver', 'SummerBronze']].sum(axis=1)
medals[medals['TotalWinterMedals'] > medals['TotalSummerMedals']][['Country', 'TotalSummerMedals', 'TotalWinterMedals']]
```

23.แสดงประเทศที่ไม่เคยได้เหรียญใดของโอลิมปิกฤดูร้อนแต่ได้เหรียญใดของฤดูหนาวหรือได้แต่เหรียญฤดูร้อนไม่เคยได้ฤดูหนาว

```python
winter_countries = medals['TotalWinterMedals'] > 0
summer_countries = medals['TotalSummerMedals'] == 0

medals[(winter_countries & summer_countries) | (~winter_countries & ~summer_countries)]
```

24.แสดงประเทศเรียงลำดับตามเหรียญทองฤดูร้อนจากน้อยไปมาก

```python
medals.sort_values(by='SummerGold', ascending=True)[['Country','SummerGold']]
```

25.แสดงประเทศเรียงลำดับตามเหรียญทองฤดูร้อนจากมากไปน้อยถ้าได้เหรียญทองฤดูร้อนเท่ากันให้เรียงตามจำนวนการเข้าแข่งฤดูร้อนจากน้อยไปมาก

```python
medals.sort_values(by=['SummerGold', 'SummerGame'], ascending=[False, True])[['Country','SummerGold', 'SummerGame']]
```

26.ให้สร้างคอลัมน์ใหม่ในตาราง medals ให้ชื่อว่า score โดยคำนวณคะแนนรวมจากเหรียญที่แต่ละประเทศได้(ทั้งฤดูร้อนและหนาว) โดยให้เหรียญทองได้ 3 คะแนนเหรียญเงินได้ 2 คะแนนเหรียญทองแดงได้ 1 คะแนนแล้วแสดงข้อมูล 10 ประเทศแรกที่ได้คะแนนเยอะที่สุด

```python
medals['score'] = medals['SummerGold']*3 + medals['SummerSilver']*2 + medals['SummerBronze']*1 + medals['WinterGold']*3 + medals['WinterSilver']*2 + medals['WinterBronze']
medals.sort_values(by='score', ascending=False).head(10)[['Country', 'score']]
```

27.การเปรียบเทียบคะแนนจากเหรียญทั้งหมดอาจจะไม่แฟร์เพราะแต่ละประเทศเข้าร่วมการแข่งขันไม่เท่ากันดังนั้นให้สร้างคอลัมน์ใหม่ในตาราง medals ชื่อว่า perGame โดยคำนวณคะแนนต่อการเข้าร่วมแข่งขัน(คะแนนคำนวณเหมือนเดิมแต่หารด้วยจำนวนเกมส์ที่เข้าแข่งขันแยกกันระหว่างฤดูร้อนกับฤดูหนาวค่าperGame จะเป็นผลบวกของคะแนนของฤดูร้อนกับฤดูหนาว) แสดงข้อมูล 10 ประเทศที่ได้คะแนนน้อยที่สุด

```python
medals['perGame'] = (medals['SummerGold']*3 + medals['SummerSilver']*2 + medals['SummerBronze']*1) / medals['SummerGame'].replace(0, 1) + \
                    (medals['WinterGold']*3 + medals['WinterSilver']*2 + medals['WinterBronze']*1) / medals['WinterGame'].replace(0, 1)

medals.sort_values(by='perGame').head(10)[['Country', 'perGame']]
```
